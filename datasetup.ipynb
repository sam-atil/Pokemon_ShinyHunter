```python
import os
import albumentations as A
import cv2
from PIL import Image
import csv

#Setting up folders
input_folder = 'data/sprite_database_mocV2'
output_folder = 'data/sprite_db'

#Creating Pipeline
transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.Transpose(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=15, p=0.7),    A.GaussNoise(var_limit=(5.0, 30.0), p=0.5),
    A.GaussianBlur(blur_limit=(1, 2), p=0.3),
    A.Perspective(scale = (0.05, 0.1), p = 0.5),
    A.CoarseDropout(max_holes=2, max_height=4, max_width=4, p=0.5)
])

#Open csv file
with open('sprite_db.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)

    #Header row
    writer.writerow(['og_file', 'filename', 'label'])


    for file_img in os.listdir(input_folder): #Returns a list of files to loop through. In our case the images

        #Grabbing file info(name, paths)
        label = os.path.splitext(file_img)[0]  # e.g., "snivy"
        pokemon_path = os.path.join(input_folder, file_img)

        #Reading in img
        img = cv2.imread(pokemon_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        #Applying transformation
        for i in range(100):
            aug_data = transform(image = img)
            aug_img = aug_data['image']
            aug_img_rgb = cv2.cvtColor(aug_img, cv2.COLOR_BGR2RGB)
            new_filename = f"{label}_{i}.png"
            output_path = os.path.join(output_folder, new_filename)

            #Saving image in folder
            cv2.imwrite(output_path, aug_img_rgb)

            #Saving image in csv file
            writer.writerow([file_img, new_filename, label])

print(f"✅ Finished augmenting all images into {output_folder}")
```

```python
import os
import albumentations as A
import cv2
from PIL import Image
import csv

#Setting up folders
input_folder = 'data/shiny_sprite_database'
output_folder = 'data/shiny_sprite_db'

#Creating Pipeline
transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.Transpose(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=15, p=0.7),    A.GaussNoise(var_limit=(5.0, 30.0), p=0.5),
    A.GaussianBlur(blur_limit=(1, 2), p=0.3),
    A.Perspective(scale = (0.05, 0.1), p = 0.5),
    A.CoarseDropout(max_holes=2, max_height=4, max_width=4, p=0.5)
])

#Open csv file
with open('shiny_sprite_db.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)

    #Header row
    writer.writerow(['og_file', 'filename', 'id', 'shiny']) #Shiny == 1, Non-Shiny == 0


    for file_img in os.listdir(input_folder): #Returns a list of files to loop through. In our case the images

        #Grabbing file info(name, paths)
        poke_name = os.path.splitext(file_img)[0]  # e.g., "snivy" or "snivy_shiny"
        pokemon_path = os.path.join(input_folder, file_img)

        print(poke_name)

        #Reading in img
        img = cv2.imread(pokemon_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        #Splitting filename to catch shiny
        split_list = poke_name.split("_") #splits filename by underscore
        print(split_list)
        #Applying transformations
        if len(split_list) == 2: #We know current file img is a shiny img
            for i in range(100):
                aug_data = transform(image = img)
                aug_img = aug_data['image']
                aug_img_rgb = cv2.cvtColor(aug_img, cv2.COLOR_BGR2RGB)
                new_filename = f"{poke_name}_{i}.png" # e.g snivy_shiny_10
                output_path = os.path.join(output_folder, new_filename)

                #Saving image in folder
                cv2.imwrite(output_path, aug_img_rgb)

                #Saving image in csv file
                #split_list[0] refers to first half of string, e.g snivy instead of snivy_shiny
                #1 refers to file img being a shiny
                writer.writerow([file_img, new_filename, split_list[0], 1]) 
            
        else: #we know current file img is not a shiny
            for i in range(100):
                aug_data = transform(image = img)
                aug_img = aug_data['image']
                aug_img_rgb = cv2.cvtColor(aug_img, cv2.COLOR_BGR2RGB)
                new_filename = f"{poke_name}_{i}.png"
                output_path = os.path.join(output_folder, new_filename)

                #Saving image in folder
                cv2.imwrite(output_path, aug_img_rgb)

                #Saving image in csv file
                #0 refers to file img not being a shiny
                writer.writerow([file_img, new_filename, poke_name, 0])

print(f"✅ Finished augmenting all images into {output_folder}")
```

```python
import os
import albumentations as A
import cv2
import csv
from tqdm import tqdm

# -----------------------------------------
# Folder paths
# -----------------------------------------

SRC_TRAIN = "Training_mon/originals/HG"   # main training sprites
SRC_VAL   = "Training_mon/originals/DP"   # validation source
SRC_TEST  = "Training_mon/originals/BL" # test source

OUT_TRAIN = "Training_mon/final/train"
OUT_VAL   = "Training_mon/final/val"
OUT_TEST  = "Training_mon/final/test"

os.makedirs(OUT_TRAIN, exist_ok=True)
os.makedirs(OUT_VAL, exist_ok=True)
os.makedirs(OUT_TEST, exist_ok=True)

# -----------------------------------------
# Augmentation Pipelines
# -----------------------------------------

train_tf = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.Transpose(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=15, p=0.7),    A.GaussNoise(var_limit=(5.0, 30.0), p=0.5),
    A.GaussianBlur(blur_limit=(1, 2), p=0.3),
    A.Perspective(scale = (0.05, 0.1), p = 0.5),
    A.CoarseDropout(max_holes=10, max_height=5, max_width=5, p=0.8),
])

val_tf = A.Compose([
    A.Resize(96,96),
    A.RandomBrightnessContrast(brightness_limit=0.05, contrast_limit=0.05, p=0.3),
    A.ShiftScaleRotate(shift_limit=0.08, scale_limit=0.08, rotate_limit=15, p=0.7)
])

test_tf = A.Compose([
    A.Resize(96,96),
    A.RandomBrightnessContrast(brightness_limit=0.05, contrast_limit=0.05, p=0.3),
    A.ShiftScaleRotate(shift_limit=0.08, scale_limit=0.08, rotate_limit=15, p=0.7)
])

N_TRAIN = 70   # e.g. 30 Pokémon * 70 = 2100
N_VAL   = 10   # 30 * 10 = 300
N_TEST  = 20   # 30 * 20 = 600

# -----------------------------------------
# Helper function to augment a folder
# -----------------------------------------
def generate_split(src_folder, out_folder, tf, count_per_class, writer):
    for filename in tqdm(os.listdir(src_folder)):
        name, ext = os.path.splitext(filename)

        # Check if shiny
        if name.endswith("_Shiny"):
            base_label = name[:-6]           # e.g., "Abra" from "Abra_Shiny"
            folder_label = base_label # folder: "Abra"
            out_name_prefix = base_label + "_Shiny"
        else:
            base_label = name                # e.g., "Abra"
            folder_label = base_label        # folder: "Abra"
            out_name_prefix = base_label

        img_path = os.path.join(src_folder, filename)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        new_out_folder = os.path.join(out_folder, folder_label)
        os.makedirs(new_out_folder, exist_ok=True)

        for i in range(count_per_class):
            aug = tf(image=img)["image"]
            aug_bgr = cv2.cvtColor(aug, cv2.COLOR_RGB2BGR)

            out_name = f"{out_name_prefix}_{i}.png"
            out_path = os.path.join(new_out_folder, out_name)

            relative_path = os.path.join(folder_label, out_name)  # path for CSV
            cv2.imwrite(out_path, aug_bgr)

            # Write CSV: relative_path, base label
            writer.writerow([relative_path, base_label])


# -----------------------------------------
# Create CSV and Sets
# -----------------------------------------
with open("Training_mon/final/train.csv", "w", newline="") as f_train:
    writer = csv.writer(f_train)
    writer.writerow(["filename", "label"])

    print("\n Generating TRAIN...")
    generate_split(SRC_TRAIN, OUT_TRAIN, train_tf, N_TRAIN, writer)

with open("Training_mon/final/val.csv", "w", newline="") as f_val:
    writer = csv.writer(f_val)
    writer.writerow(["filename", "label"])

    print("\n Generating VAL...")
    generate_split(SRC_VAL, OUT_VAL, val_tf, N_VAL, writer)


with open("Training_mon/final/test.csv", "w", newline="") as f_test:
    writer = csv.writer(f_test)
    writer.writerow(["filename", "label"])

    print("\n Generating TEST...")
    generate_split(SRC_TEST, OUT_TEST, test_tf, N_TEST, writer)

print("\n Done!✅")
```
